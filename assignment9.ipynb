# 클래스 이용 & LM 이용 방안 고민

```python
import math

class ngram_prob:
    def __init__(self):
        self.ls = []
        self.freq = dict()

        self.token_sum = 0       # 총 토큰 수
        self.unigram_sum = 0     # word의 총 유니그램 수
        self.bigram_sum = 0      # 총 바이그램 수 (어절 단위)
        self.trigram_sum = 0     # 총 트라이그램 수 (어절 단위)

    def openfile(self, file, word, n):
        with open(file, encoding = 'utf8') as fo:
            for In, line in enumerate(fo,1):
                line = line.strip()
                token = line.split(' ')

                self.token_sum += len(token)
                self.bigram_sum += len(token) - 1
                self.trigram_sum += len(token) - 2

                if word in token:
                    self.unigram_sum += token.count(word)
                    new_token = token[token.index(word):]

                    # word로 시작하는 n-gram 추출 및 리스트 ls 담기
                    for i in range(n):
                        for j in range(len(new_token) - i):
                            if new_token[j] != word: continue
                            self.ls.append(new_token[j:j+i+1])

        # word로 시작하는 모든 ngram에 대한 빈도 사전 만들기
        for ngram in self.ls:
            if tuple(ngram) not in self.freq:
                self.freq[tuple(ngram)] = 0
            self.freq[tuple(ngram)] += 1

        # 해당 ngram이 
        for k in self.freq:
            if len(k) == 1:    # 유니그램이면
                self.freq[k] = math.log(self.freq[k]/self.token_sum)
            elif len(k) == 2:  # 바이그램이면
                self.freq[k] = math.log(self.freq[k]/self.bigram_sum)
            elif len(k) == 3:  # 트라이그램이면
                self.freq[k] = math.log(self.freq[k]/self.trigram_sum)

        for x,y in sorted(self.freq.items(), key = lambda x: (-x[1], x[0])):
            print(y, x, sep='\t')
```

```python
import math

token_sum = 0
unigram_sum = 0
bigram_sum = 0
trigram_sum = 0

n = 3
ls = []
freq = dict()

with open('05.corpus.txt', encoding = 'utf8') as file:
    for In, line in enumerate(file,1):
        #print(In, line, sep='\t')
        line = line.strip()
        token = line.split(' ')
        
        token_sum += len(token)
        bigram_sum += len(token) - 1
        trigram_sum += len(token) - 2
        
        if '학교에' in token:
            unigram_sum += token.count('학교에')
            new_token = token[token.index('학교에'):]
            
            for i in range(n):
                for j in range(len(new_token) - i):
                    if new_token[j] != '학교에': continue
                    ls.append(new_token[j:j+i+1])

print(token_sum) # 총 698,148 토큰
print(unigram_sum) # '학교에' : 97개
print(bigram_sum) # 총 598,148 바이그램
print(trigram_sum) # 총 498,148 트라이그램

print(math.log(unigram_sum/token_sum)) # -8.881475415148968 --> 정답과 조금 다름
#print(ls)

for i in ls:
    if tuple(i) not in freq:
        freq[tuple(i)] = 0
    freq[tuple(i)] += 1

#print(freq)

for x,y in sorted(freq.items(), key = lambda x: (-x[1], x[0])):
    print(y,x,math.log(y/token_sum),sep='\t')
```
